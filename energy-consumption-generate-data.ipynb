```python
import pandas as pd
import numpy as np

# Step 1: Generate timestamps for 1 year, hourly
date_rng = pd.date_range(start='2023-01-01', end='2023-12-31 23:00:00', freq='H')

# Step 2: Departments
departments = ['Production', 'Admin', 'IT', 'HR', 'R&D']

# Step 3: Cost per unit
electricity_unit_cost = 10  # ₹ per kWh
water_unit_cost = 2  # ₹ per 100 liters

# Step 4: Generate dataset
data = []

np.random.seed(42)  # for reproducibility

for ts in date_rng:
    for dept in departments:
        # base consumption per dept
        if dept == 'Production':
            elec = np.random.normal(120, 20)  # kWh
            water = np.random.normal(500, 50)  # liters
        elif dept == 'Admin':
            elec = np.random.normal(50, 10)
            water = np.random.normal(200, 30)
        elif dept == 'IT':
            elec = np.random.normal(60, 15)
            water = np.random.normal(150, 20)
        elif dept == 'HR':
            elec = np.random.normal(40, 10)
            water = np.random.normal(100, 15)
        else:  # R&D
            elec = np.random.normal(80, 20)
            water = np.random.normal(250, 30)

        # Introduce random spikes for anomalies
        if np.random.rand() < 0.01:  # 1% chance of spike
            elec *= np.random.uniform(1.5, 2)
            water *= np.random.uniform(1.5, 2)

        elec_cost = elec * electricity_unit_cost
        water_cost = (water/100) * water_unit_cost
        total_cost = elec_cost + water_cost

        data.append([ts, dept, round(elec,2), round(water,2), round(elec_cost,2), round(water_cost,2), round(total_cost,2)])

# Step 5: Create DataFrame
df = pd.DataFrame(data, columns=['timestamp','department','electricity_kWh','water_liters','electricity_cost','water_cost','total_cost'])

# Step 6: Save to CSV for backup
df.to_csv('energy_consumption.csv', index=False)

df.head(10)
```
